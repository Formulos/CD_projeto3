```python
def minimos_quadrados(x,y): #função que calcula a regressão linear, so funciona para dois eixos
    
    x = sm.add_constant(x)
    est = sm.OLS(y, x)
    est = est.fit()

    X_prime = np.linspace(Data_Total.Label.min(), Data_Total.Label.max(), 1000)[:, np.newaxis]
    X_prime = sm.add_constant(X_prime)  # add constant as we did before
    y_hat = est.predict(X_prime)
    summary = est.summary()
    
    return X_prime,y_hat,summary
```

```python
label = pd.read_excel('indicator_t above 15 labour to pop percent.xlsx') #Taxa de população economicamente ativa
government = pd.read_excel('government.xlsx') #Per capita government expenditure on health at average exchange rate (US$)
expenditure = pd.read_excel('indicator_f above 15 employ.xlsx') #Females 15-64 labour forçe participation(%)

# Todos os dados são referentes ao ano de 2007
```

```python
label = pd.DataFrame(label.year[label.year>=0])
government = pd.DataFrame(government.year[government.year>=0])
expenditure = pd.DataFrame(expenditure.year[expenditure.year>=0])
```

```python
final = label.join(government, lsuffix='_label')
final = final.join(expenditure,lsuffix='_government', rsuffix='_expenditure')
```

```python
final = final.dropna()
final.columns=['Label','Government','Expenditure']
final.head()
```

```python
Data_Total = final

f, ax = plt.subplots(1,2,figsize=(16,8))
ax[0].scatter(Data_Total.Government, Data_Total.Label, alpha=0.7)  # Plot the raw data
ax[0].set_xlabel("Gasto com saude")
ax[0].set_ylabel("Pea")

y = Data_Total.Label #variavel resposta
x = Data_Total.Government

print("Correlação entre Gasto com saude e Pea:" ,np.corrcoef(x,y)[0][1])


X_prime,y_hat,summary_G = minimos_quadrados(x,y)
ax[0].plot(X_prime[:, 1], y_hat, 'r', alpha=0.9) 

ax[1].scatter(Data_Total.Expenditure, Data_Total.Label, alpha=0.7)  # Plot the raw data
ax[1].set_xlabel("Participação de mulheres no mercado de trabalho")
ax[1].set_ylabel("Pea")

x = Data_Total.Expenditure

print("Correlação entre a Participação de mulheres no mercado de trabalho e o Pea:",np.corrcoef(x,y)[0][1])
X_prime,y_hat,summary_E = minimos_quadrados(x,y)
ax[1].plot(X_prime[:, 1], y_hat, 'r', alpha=0.9) 

plt.tight_layout(1)



```

Como podemos ver a correlação entre o pea e a Participação das mulheres é alta (0.85), Porem a correlação entre o pea e o gasto com saude é baixo.

```python
# calculo da curva de regressão de 3 dimensões

X=Data_Total.iloc[:,(1,2)]

X=sm.add_constant(X)  # Adds a constant term to the predictor
X.head()

model=sm.OLS(y, X)
fit=model.fit()
fit.summary()
```

```python


fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')
ax.scatter(Data_Total.Expenditure,Data_Total.Government, Data_Total.Label,c = 'r')
```

```python

final_log = Data_Total

X = sm.add_constant(X)
est = sm.OLS(y, X).fit()

xx1, xx2 = np.meshgrid(np.linspace(final_log.Label.min(), final_log.Label.max(), 100), 
                       np.linspace(final_log.Expenditure.min(), final_log.Expenditure.max(), 100))

Z = est.params[0] + est.params[1]*xx2 + est.params[2]*xx1 

fig = plt.figure(figsize=(12, 8))
ax = Axes3D(fig, azim=-115, elev=15)

# plotando hiperplano, azul significa menos dollares gastos com saude, vermelho é mais dollares gastos com saude
surf = ax.plot_surface(xx1, xx2, Z, cmap=plt.cm.RdBu_r, alpha=0.6, linewidth=0)

# plotando pontos - acima do hiperplano sao brancos, abaixo sao pretos
resid = y - est.predict(X)
ax.scatter(final_log[resid >= 0].Label, final_log[resid >= 0].Expenditure, y[resid >= 0], c="white", alpha=1.0, s=50)
ax.scatter(final_log[resid < 0].Label, final_log[resid < 0].Expenditure, y[resid < 0], c="black", alpha=1.0, s=50)

ax.set_xlabel('Pea')
ax.set_ylabel('Participação de mulheres no mercado de trabalho')
ax.set_zlabel('Gasto com Saude')
```

Como podemos ver nos conseguimos obter uma curva de regresão que se enquadra bem com os nossos dados, porem já que uma das variaveis tem uma correlação fraca com a Pea um resultado semelhante poderia ser obtido se fossem usados somente o Pea a participação de mulheres, Para melhorarmos a curva de regressão seria necessario melhorar os dados.
